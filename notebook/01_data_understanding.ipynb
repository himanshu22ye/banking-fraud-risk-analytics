## 📄 Column Descriptions (Accounts Dataset)

**Bank Name:** Name of the financial institution where the account is maintained.

**Bank ID:** Unique numerical identifier assigned to each bank.

**Account Number:** Unique identifier representing a customer's bank account within a bank.

**Entity ID:** Unique identifier assigned to the account holder (individual or organization).

**Entity Name:** Name of the account owner or business entity, such as a Corporation, Partnership, or Sole Proprietorship.

## 📄 Column Descriptions (Transactions Dataset)

**Timestamp:** Date and time when the transaction was initiated.

**From Bank:** Unique identifier of the bank sending the funds.

**Account:** Account number of the sender initiating the transaction.

**To Bank:** Unique identifier of the bank receiving the funds.

**Account.1:** Account number of the recipient receiving the transaction.

**Amount Received:** Amount credited to the recipient's account in the receiving currency.

**Receiving Currency:** Currency in which the recipient receives the funds.

**Amount Paid:** Amount debited from the sender's account in the payment currency.

**Payment Currency:** Currency used by the sender to make the payment.

**Payment Format:** Method or channel through which the transaction was processed (e.g., Cheque, Reinvestment, Credit Card, Cash, Wire Transfer).

**Is Laundering:** Binary target variable indicating whether the transaction is associated with money laundering.
- **0:** Legitimate Transaction
- **1:** Suspicious / Money Laundering Transaction
