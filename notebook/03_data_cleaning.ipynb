## Droping the dupilcates

```python
transactions_clean = transactions_clean.drop_duplicates()
```

```python
transactions_clean.duplicated().sum()
```

## Duplicate Record Handling

During the data quality assessment, the Transactions dataset was found to contain **9 duplicate records**.

A detailed investigation was performed by displaying both the original and duplicate entries using `duplicated(keep=False)`.

The comparison confirmed that each duplicate pair was **identical across all 11 columns**, indicating that these were duplicate observations rather than legitimate repeated transactions.

To maintain data integrity and prevent double-counting during analysis, the duplicate records were removed from the dataset.

The Accounts dataset did not contain any duplicate records.

## Standardize Column Names

To improve readability, maintain consistency, and simplify coding, the column names are standardized using the **snake_case** naming convention.

This makes the dataset easier to work with in Python, SQL, and Power BI while following common data engineering and analytics best practices.

```python
transactions_clean.rename(
    columns={
        "Timestamp": "timestamp",
        "From Bank": "from_bank",
        "Account": "sender_account",
        "To Bank": "to_bank",
        "Account.1": "receiver_account",
        "Amount Received": "amount_received",
        "Receiving Currency": "receiving_currency",
        "Amount Paid": "amount_paid",
        "Payment Currency": "payment_currency",
        "Payment Format": "payment_format",
        "Is Laundering": "is_laundering"
    },
    inplace=True
)
```

```python
account_clean.rename(
    columns={
        "Bank Name": "bank_name",
        "Bank ID": "bank_id",
        "Account Number": "account_number",
        "Entity ID": "entity_id",
        "Entity Name": "entity_name"
    },
    inplace=True
)
```

```python
print("Account columns :",account_clean.columns)
print("Transactions columns :",transactions_clean.columns)
```

```python
transactions_clean["timestamp"] = pd.to_datetime(
    transactions_clean["timestamp"]
)
```
